```python
def grouped_mean_heatmap(x1, x2, y):
    group = df.groupby([x1, x2])[y].mean()    
    sns.heatmap(group.unstack(), cmap='RdBu')
    plt.ylabel(x1)
    plt.xlabel(x2)
    plt.title('Average {}'.format(y))
    plt.show()
```
